## Prérequis et notations

### Notations

On note en minuscule souligné les vecteurs, en calligraphié les
opérateurs linéaires. On note en gras les grandeurs exprimées dans une
base : en majuscule les matrices, en minuscule les vecteurs.
${\mathbf{A}}$ est la matrice de coefficients $(a_{ij})$.
${\mathbf{A}}$ est la représentation de $\mathcal{A}$ dans
une base (sauf précision contraire, la base canonique) tout comme
${\mathbf{v}}$ est la représentation du vecteur
${\underline{v}}$, ses composantes sont les $(v_i)$. On
passera implicitement d'une famille de vecteurs colonnes
$({\mathbf{a}}_i)$ à une matrice :
${\mathbf{A}}=[{\mathbf{a}}_1,\ldots{\mathbf{a}}_n]$.

Par défaut l'espace vectoriel ${\mathbb{E}}$ sur le corps
$\mathbb{K}$ (en pratique ce sera ${\mathbb{R}}$ ou ${\mathbb{C}}$)
est de dimension $n$, les matrices sont de dimension $m\times n$ ($m$
lignes, $n$ colonnes). Les bornes des indices sont omises quand il n'y a
pas de risque de confusion. Les vecteurs $({\mathbf{e}}_i)$
forment la base orthonormale canonique ($e_{ij}=\delta_{ij}$ symbole de
Kronecker).

En général, les propriétés sont énoncées dans ${\mathbb{C}}$
et il est précisé si elles restent valables dans
${\mathbb{R}}$.

Certains des rappels présentés ici seront mis en oeuvre sur la base de quelques exemples.

**Définition 2.1** (Matrice symétrique, hermitienne, normale, unitaire,
orthogonale.). Une matrice ${\mathbf{A}}$ est :

-   hermitienne si ${\mathbf{A}}^H={\mathbf{A}}$ ;

-   anti-hermitienne si
    ${\mathbf{A}}^H=-{\mathbf{A}}$ ;

-   unitaire si
    ${\mathbf{A}}{\mathbf{A}}^H={\mathbf{A}}^H{\mathbf{A}}={\mathbf{I}}$
    ;

-   normale si
    ${\mathbf{A}}{\mathbf{A}}^H={\mathbf{A}}^H{\mathbf{A}}$

-   symétrique si ${\mathbf{A}}$ est réelle et
    ${\mathbf{A}}={\mathbf{A}}^T$ ;

-   orthogonale si ${\mathbf{A}}$ est réelle et
    ${\mathbf{A}}{\mathbf{A}}^T={\mathbf{A}}^T{\mathbf{A}}={\mathbf{I}}$ ;

**Proposition 2.1**. *Une matrice se décompose de manière unique en
parties hermitienne et anti-hermitienne
:$$\displaystyle{\mathbf{A}}=\frac{{\mathbf{A}}+{\mathbf{A}}^H}{2}+\frac{{\mathbf{A}}-{\mathbf{A}}^H}{2}$$*

On propose de programmer quelques fonctions qui permettent de :
- déterminer le rayon spectral d'une matrice,
- vérifier que la trace d'une matrice est égale à la somme de ses valeurs propres et que son déterminant est le produit de ses valeurs propres, on comparera avec le calcul de la trace et du déterminant par numpy,
- tester les propriétés d'une matrice : symétrique, antisymétrique, orthogonale,

On donne ci-dessous des matrices pour tester ces fonctions.

```python
import numpy as np
from numpy import linalg as la
n=5
# A est une matrice symétrique, définie et positive
A = np.random.random((n, n))
A=A*A.transpose()+n*np.identity(n)
# B est une matrice orthogonale
H = np.random.random((n, n))
u, s, vh = np.linalg.svd(H, full_matrices=False)
B = u.dot(vh)
# C est définie comme B A BT (C et A sont la même matrice dans 2 bases différentes)
C = B.dot(A.dot(np.transpose(B)))
# D est une matrice quelconque
D = np.random.random((n, n))

def rayon_spectral(A):
    Rs = la.eigvals(A)
    return np.max(abs(Rs))

def verif_sum_trace(A):
    Tr1 = A.trace()
    Tr2 = sum(la.eigvals(A))
    return np.allclose(Tr1, Tr2, 1E-16)
# Le all close permet de verifier si deux vals sont proches ou non par rapp à une tolérance
# On peut mettre rtol = 1E-16 comme relative ou atol = ... pour erreur absolue

def verif_mult_trace(A):
    det1 = la.det(A)
    det2 = np.real_if_close(np.prod(la.eigvals(A)))
    return np.allclose(det1, det2, 1E-16)

def is_symmetric(A):
    return np.allclose(A, A.T, 1E-16)

def is_antisymetric(A):
    return np.allclose(A.T, -A, 1E-16)

def is_orthogonal(A):
# Est orthogonale si A'A = AA' = I
   return np.allclose((A@A.T), (A.T@A), 1E-16) and np.allclose((A@A.T), np.identity(len(A)), 1E-16)

# A, B, C, D

print('Rayon spectral de A = ', rayon_spectral(A), 'Rayon spectral de B = ', rayon_spectral(B), 'Rayon spectral de C = ', rayon_spectral(C), 'Rayon spectral de D = ', rayon_spectral(D), )
print('Verification avec la somme de la trace :', verif_sum_trace(A),'; ', verif_sum_trace(B),'; ', verif_sum_trace(C),'; ', verif_sum_trace(D))
print('Verification de la multiplication des eigvals :' ,verif_mult_trace(A),'; ',verif_mult_trace(B),'; ',verif_mult_trace(C),'; ',verif_mult_trace(D) )
print('Est symétrique? ', is_symmetric(A),'; ', is_symmetric(B),'; ', is_symmetric(C),'; ', is_symmetric(D))
print('Est antisymétrique? ', is_antisymetric(A),'; ', is_antisymetric(B),'; ', is_antisymetric(C),'; ', is_antisymetric(D))
print('Est orthogonale? ', is_orthogonal(A),'; ', is_orthogonal(B),'; ', is_orthogonal(C),'; ', is_orthogonal(D))
```

**Définition 2.2**. À une matrice carrée ${\mathbf{A}}$ on
peut associer une forme quadratique :
${\mathbf{v}}\mapsto {\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}$.



**Proposition 2.2**. *Une matrice hermitienne définit une forme
quadratique réelle.*

*Une matrice anti-hermitienne définit une forme quadratique purement
imaginaire.*

*Une matrice antisymétrique définit une forme quadratique nulle sur
${\mathbb{R}}$*


*Preuve* Soit ${\mathbf{v}}\in{\mathbb{E}}$, et
${\mathbf{A}}$ une matrice carrée
${\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}=\left({\mathbf{A}}^H{\mathbf{v}}\right)^H{\mathbf{v}}=\overline{{\mathbf{v}}^H{\mathbf{A}}^H{\mathbf{v}}}$

Si ${\mathbf{A}}$ est hermitienne alors
${\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}=\overline{{\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}}$
donc la forme est réelle

Si ${\mathbf{A}}$ est anti-hermitienne alors
${\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}=-\overline{{\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}}$
donc la forme est purement imaginaire

Si ${\mathbf{A}}$ est anti-symétrique alors
${\mathbf{v}}^T{\mathbf{A}}{\mathbf{v}}=-{{\mathbf{v}}^T{\mathbf{A}}{\mathbf{v}}}$
donc la forme est nulle. ◻

**Définition 2.3** (Matrice positive, matrice définie, ordre partiel de
Löwner). Soit une matrice carrée $\mathbf{A}$.

Elle est positive si et seulement si la forme quadratique qui lui est
associée est réelle positive
$\forall{\mathbf{v}}\in{\mathbb{E}},\ {\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}}\geqslant 0$.

Elle est définie si et seulement si
${\mathbf{v}}^H{\mathbf{A}}{\mathbf{v}} = 0\Leftrightarrow {\mathbf{v}} = 0$.

On peut munir l'ensemble des matrices hermitiennes d'un ordre partiel
(de Löwner) :
${\mathbf{A}}\leqslant{\mathbf{B}}\Leftrightarrow \forall {\mathbf{x}}\in{\mathbb{E}}, {\mathbf{x}}^H{\mathbf{A}}{\mathbf{x}}\leqslant{\mathbf{x}}^H{\mathbf{B}}{\mathbf{x}}$.
La positivité d'une matrice au sens de Löwner correspond à la positivité
de la forme quadratique.


*Remarque 2.2*. Il existe une autre notion d'ordre partiel pour les
matrices réelles (pas nécessairement symétriques) :
${\mathbf{A}}\geqslant{\mathbf{B}} \Leftrightarrow \forall (i,j),\  a_{ij}\geqslant b_{ij}$.
Une matrice positive est alors une matrice dont tous les coefficients
sont positifs.

**Définition 2.4** (Matrice à diagonale dominante.). Une matrice
${\mathbf{A}}$ est à diagonale dominante si
$\forall i,\ |a_{ii}|\geqslant\sum_{j\neq i}|a_{ij}|$.


## Réduction des matrices

### Résultats généraux de matrices équivalentes

La *réduction* consiste à écrire une application dans la base qui rend
la matrice associée la plus simple possible.


**Théorème 2.1**. *Principaux résultats de réduction :*

-   *Étant donné une matrice *carré* $\mathbf{A}$, il existe une matrice
    *unitaire* $\mathbf{U}$ telle que la matrice
    ${\mathbf{T}}={\mathbf{U}}^{-1}{\mathbf{A}}{\mathbf{U}}$
    soit *triangulaire* (théorème de trigonalisation de Schur).*

-   *Étant donné une matrice *normale* $\mathbf{A}$, il existe une
    matrice *unitaire* $\mathbf{U}$ telle que la matrice
    ${\mathbf{D}}={\mathbf{U}}^{-1}{\mathbf{A}}{\mathbf{U}}$
    soit *diagonale*.*

-   *Étant donné une matrice *hermitienne* (respectivement *symétrique*)
    $\mathbf{A}$, il existe une matrice *unitaire* (respectivement
    *orthogonale*) $\mathbf{U}$ telle que la matrice
    ${\mathbf{D}}={\mathbf{U}}^{-1}{\mathbf{A}}{\mathbf{U}}$
    soit *diagonale réelle*.*

**Proposition 2.3**. 

-   *Les éléments diagonaux de $\mathbf{T}$ ou $\mathbf{D}$ sont les
    valeurs propres de ${\mathbf{A}}$.*

-   *Les valeurs propres d'une matrice unitaire ou orthogonale sont des
    nombres complexes de module $1$.*


*Remarque 2.3*. Pour mémoire, la matrice semblable la plus simple pour
une matrice quelconque s'écrit sous la forme de Jordan (bloc-diagonale
avec des $1$ sur la première surdiagonale).



### Décomposition en valeurs singulières

Un problème plus général consiste à trouver l'expression d'une *matrice
équivalente* particulièrement simple. Dans ce cas là l'application n'est
plus nécessairement un endomorphisme,
$\mathcal{A}:{\mathbb{E}}\to \mathbb{F}$, les matrices
associées peuvent être rectangulaires et on introduit des changements de
base dans ${\mathbb{E}}$ et $\mathbb{F}$. On a donc
$\mathbf{Q}$ isomorphisme de $\mathbb{F}$, $\mathbf{P}$ isomorphisme de
${\mathbb{E}}$ et
$${\mathbf{B}}={\mathbf{Q}}{\mathbf{A}}{\mathbf{P}}$$

**Théorème 2.2**. *Décomposition en valeurs singulières (SVD) : si
$\mathbf{A}$ est une matrice complexe $m\times n$ (respectivement
réelle), il existe deux matrices unitaires (respectivement orthogonales)
$\mathbf{U}$ ($n\times n$) et $\mathbf{V}$ ($m\times m$) telles que
$${\mathbf{U}}^H {\mathbf{A}} {\mathbf{V}} = {\mathbf{\Sigma}} \qquad \qquad  {\mathbf{\Sigma}}= \begin{pmatrix}
    \sigma_1 & 0 & \cdots & 0   \\
    0 & \ddots & \ddots & \vdots \\ 
    \vdots & \ddots & \ddots & 0 \\
    \vdots & & \ddots &  \sigma_m\\
    \vdots &  &  &  0 \\
    0 & \cdots & \cdots & 0
    \end{pmatrix}\qquad   \text{matrice ``diagonale'' } m\times n$$ où
l'on a choisi de représenter le cas $n>m$. Les $\sigma_i$ sont appelées
valeurs singulières de ${\mathbf{A}}$, ce sont les racines
carrées des valeurs propres de la matrice
${\mathbf{A}}^H{\mathbf{A}}$, elles sont
nécessairement *réelles, positives ou nulles* ; on peut choisir de les
disposer dans n'importe quel ordre, pour un ordre donné la décomposition
est unique. Par la suite on choisit l'ordre décroissant et alors
$\sigma_1=\sqrt{\rho({\mathbf{A}}^H{\mathbf{A}})}$.*

*Remarque 2.4*. Le résultat s'écrit aussi :
$${\mathbf{A}}  = {\mathbf{U}}{\mathbf{\Sigma}}{\mathbf{V}}^H = \sum_i \sigma_i {\mathbf{u}}_i {\mathbf{v}}_i^H$$
ce qui décompose la matrice en la somme de
$\operatorname{rg}({\mathbf{A}})$ contributions de rang $1$.


*Remarque 2.5*. Quitte à se répéter, la SVD d'une matrice réelle ne fait intervenir que des matrices réelles.

Nous allons coder la SVD dans le cas d'une matrice carrée réelle. On écrit donc ${\mathbf{A}}  = {\mathbf{U}}{\mathbf{\Sigma}}{\mathbf{V}}^T$
Le calcul de la SVD consiste à trouver les valeurs propres et les vecteurs propres de $\mathbf{A}\mathbf{A}^T$ et $\mathbf{A}^T\mathbf{A}$. Les vecteurs propres de $\mathbf{A}^T\mathbf{A}$ constituent les colonnes de $\mathbf{V}$, les vecteurs propres de $\mathbf{A}\mathbf{A}^T$ constituent les colonnes de $\mathbf{U}$. Et comme indiqué, les valeurs singulières de $\mathbf{\Sigma}$ sont des racines carrées des valeurs propres de $\mathbf{A}\mathbf{A}^T$ ou $\mathbf{A}^T\mathbf{A}$. 
Les valeurs singulières sont les entrées diagonales de la matrice $\mathbf{\Sigma}$ et sont classées par ordre décroissant. Comme ici la matrice $\mathbf{A}$ est une matrice réelle, alors $\mathbf{U}$ et $\mathbf{V}$ sont également réels. Enfin, il ne faut pas oublier qu'on doit avoir : $\mathbf{A}\mathbf{V} = \mathbf{U}\mathbf{\Sigma}$


```python
%config Completer.use_jedi = False


import numpy as np 
from numpy import linalg as la
n = 20
A = np.random.random((n,n))
#A = np.array([[3,2,2],[2,3,-2]])


def svd_mine(A):
    eAAT, U = la.eigh(A.dot(np.transpose(A)))
    eATA, V = la.eigh(np.transpose(A).dot(A))

    
    SigX = np.sqrt(eAAT)
    idx = SigX.argsort()[::-1]
    SigX = SigX[idx]
    S = np.diag(SigX)



    U = U[:, idx]
    V = V[:, idx]
    
    AV = A.dot(V)
    US=U.dot(S)
    sign = np.isclose(AV, US, rtol = 1E-16).astype(int)-np.logical_not(np.isclose(AV, US, rtol = 1E-16)).astype(int)
    #print(V)
    #print(sign)
    #print(np.multiply(sign,V))
    V = np.multiply(sign,V)
    svdA = U@S@V.T
    return svdA, S, SigX, U, V


svdA, Sig, SigX, U, V = svd_mine(A)
Unp, Snp, Vnp = la.svd(A)
svdAP = Unp@np.diag(Snp)@Vnp


print(np.isclose(A, svdA, rtol = 1E-16).all())
print(np.isclose(svdA, svdAP, rtol = 1E-16).all())



```

**Proposition 2.4**. *Si
$\sigma_1\geqslant \cdots\geqslant \sigma_r > \sigma_{r+1}=\cdots=\sigma_n=0$
on a : $$\begin{aligned}
    {\operatorname{rg}({\mathbf{{\mathbf{A}}}})}&={\operatorname{rg}({\mathbf{{\mathbf{A}}^H}})}=r & \dim({\operatorname{Ker}({\mathbf{{\mathbf{A}}}})})&=m-r,\ \dim({\operatorname{Ker}({\mathbf{{\mathbf{A}}^H}})})=n-r\\
    {\operatorname{Ker}({\mathbf{{\mathbf{A}}}})}&={\operatorname{vect}}({\mathbf{v}}_{r+1},\cdots,{\mathbf{v}}_{m})&{\operatorname{Ker}({\mathbf{{\mathbf{A}}^H}})}&={\operatorname{vect}}({\mathbf{u}}_{r+1},\cdots,{\mathbf{u}}_{n})\\
    {\operatorname{Im}({\mathbf{{\mathbf{A}}}})}&={\operatorname{vect}}({\mathbf{u}}_{1},\cdots,{\mathbf{u}}_{r})&{\operatorname{Im}({\mathbf{{\mathbf{A}}^H}})}&={\operatorname{vect}}({\mathbf{v}}_{1},\cdots,{\mathbf{v}}_{r})
    \end{aligned}$$*

La SVD est donc une transformation qui apporte énormément d'information,
elle est peut être obtenue par des algorithmes numériquement très
stables mais très coûteux.

Une valeur singulière très faible (ou nulle) indique que la matrice
n'est pas inversible en pratique (matrice de rang numérique déficient).
Le résultat suivant connecte la SVD à la distance d'une matrice aux
matrices de rang déficient (la norme $2$ des matrices est étudiée plus tard).


**Proposition 2.5**.
*$$\min_{{\operatorname{rg}({\mathbf{B}})}=k}\|{\mathbf{A}}-{\mathbf{B}}\|_2 = \|{\mathbf{A}}-{\mathbf{A}}_k\|_2 = \sigma_{k+1}\quad\text{ où } {\mathbf{A}}_k=\sum_{i=1}^k\sigma_i{\mathbf{u}}_i {\mathbf{v}}_i^H$$*

Cette équation montre que la distance d'une matrice aux matrices de rang
$k$ est caractérisée par la $(k+1)$-ème valeur singulière : en
particulier, il suffit d'une perturbation de l'ordre de $\sigma_n>0$
pour rendre une matrice non-inversible. On voit aussi que
${\mathbf{A}}_k$ est le meilleur représentant de
${\mathbf{A}}$ de rang $k$ (au sens de la norme euclidienne),
cette propriété est à la base des techniques de compression de données
et de réduction de modèle.

Sur la base de la norme $2$, nous allons évaluer la représentativité des ${\mathbf{A}}$ de rang $k$.
Dans un premier temps, on trace les valeurs singulières obtenues par la SVD.
Dans un second temps, on trace la norme $2$ de $({\mathbf{A}}$-${\mathbf{A}}_k)$.

## Find what are the bird and schewefel functions

```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

# La fonction bird
def bird(x,y):
 return np.sin(x)*(np.exp(1-np.cos(y))**2)+np.cos(y)*(np.exp(1-np.sin(x))**2)+(x-y)**2

x= np.linspace(-2*math.pi,2*math.pi,100)
y= np.linspace(-2*math.pi,2*math.pi,100)
X,Y= np.meshgrid(x,y)
Z= bird(X,Y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot_surface(X,Y,Z,cmap='coolwarm')
plt.show()
```

```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

# La fonction schwefel
def schwefel(x,y):  
    return 418.9829*2 - x * np.sin(np.sqrt(abs( x )))-y*np.sin(np.sqrt(abs(y)))

x=np.linspace(-500,500)
y=np.linspace(-500,500)

X,Y=np.meshgrid(x,y)
Z=schwefel(X,Y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot_surface(X,Y,Z,cmap='jet')
plt.show()
```

### Quotient de Rayleigh des matrices symétriques et hermitiennes


**Définition 2.5** (Quotient de Rayleigh.). Soit
${\mathbf{A}}$ une matrice hermitienne, son quotient de
Rayleigh est la fonction :
$${\mathbf{v}}\neq 0\in{\mathbb{E}},\qquad R_{{\mathbf{A}}}({\mathbf{v}})=\frac{{\mathbf{v}}^H {\mathbf{A}} {\mathbf{v}}}{{\mathbf{v}}^H {\mathbf{v}}}\ \in {\mathbb{R}}$$

On remarque que
$R_{{\mathbf{A}}}(\alpha {\mathbf{v}})=R_{{\mathbf{A}}}({\mathbf{v}})$,
donc on peut travailler sur la sphère unité euclidienne
$({\mathbf{v}}^H {\mathbf{v}}=1)$.

**Proposition 2.6**. *Si les valeurs propres sont ordonnées
$\lambda_1\geqslant\cdots\geqslant\lambda_n$ et qu'on note
${\mathbf{p}}_i$ les vecteurs propres orthonormés (en norme
2) associés. Si
${\mathbb{E}}_k = {\operatorname{vect}}({\mathbf{p}}_1,\cdots,{\mathbf{p}}_k)$,
et $\mathcal{E}_k$ est l'ensemble des sous-espaces de dimension $k$, on
a :*

-   *$\lambda_k=R_{{\mathbf{A}}}({\mathbf{p}}_k)$*

-   *$\lambda_k=\min_{{\mathbf{v}}\in {\mathbb{E}}_k}R_{{\mathbf{A}}}({\mathbf{v}})$*

-   *$\lambda_k=\max_{{\mathbf{v}} \perp {\mathbb{E}}_{k-1}}R_{{\mathbf{A}}}({\mathbf{v}})$*

-   *$\lambda_k=\max_{\mathbb{V}\in\mathcal{E}_k}\min_{{\mathbf{v}} \in \mathbb{V}}R_{{\mathbf{A}}}({\mathbf{v}})$*

-   *$\lambda_k=\min_{\mathbb{V}\in\mathcal{E}_{k-1}}\max_{{\mathbf{v}} \perp \mathbb{V}}R_{{\mathbf{A}}}({\mathbf{v}})$*

-   *$\{R_{{\mathbf{A}}}({\mathbf{v}}), {\mathbf{v}}\in {\mathbb{E}}\}=[\lambda_1,\lambda_n]\subset{\mathbb{R}}$*

**Théorème 2.3** (d'entrelacement de Cauchy). *Si
${\mathbf{H}}$ est une matrice hermitienne de taille $n$ et
${\mathbf{A}}$ une sous-matrice principale de
${\mathbf{H}}$ de taille $r$, alors les valeurs propres de
$A$ séparent celles de ${\mathbf{H}}$ et plus précisément
pour $1\leqslant i\leqslant r$ :
$$\lambda_i({\mathbf{H}})\geqslant\lambda_i({\mathbf{A}})\geqslant \lambda_{n-(r-i)}({\mathbf{H}})$$*


## Normes

### Normes vectorielles


**Définition 2.6** (p-norme). La classe la plus utilisée de normes
vectorielles est celle des $p$-normes :
$$\|{\mathbf{x}}\|_p= (|x_1|^p+\cdots+|x_n|^p)^{1/p}$$


**Théorème 2.4** (Inégalité de Holder et de Cauchy-Schwartz). *:
$$\begin{aligned}
    |{\mathbf{x}}^H {\mathbf{y}}|&\leqslant \|{\mathbf{x}}\|_p\|{\mathbf{y}}\|_q \qquad \frac{1}{p}+\frac{1}{q}=1 \\
    |{\mathbf{x}}^H {\mathbf{y}}|&\leqslant \|{\mathbf{x}}\|_2\|{\mathbf{y}}\|_2
    \end{aligned}$$*



**Proposition 2.7**. *On sait que toutes les normes sont équivalentes,
les encadrements suivants sont optimaux : $$\begin{aligned}
    \|{\mathbf{x}}\|_2      &\leqslant \|{\mathbf{x}}\|_1 \leqslant \sqrt{n} \|{\mathbf{x}}\|_2\\
    \|{\mathbf{x}}\|_\infty &\leqslant \|{\mathbf{x}}\|_2 \leqslant \sqrt{n} \|{\mathbf{x}}\|_\infty\\
    \|{\mathbf{x}}\|_\infty &\leqslant \|{\mathbf{x}}\|_1 \leqslant n \|{\mathbf{x}}\|_\infty
    \end{aligned}$$*


### Normes matricielles

**Définition 2.7** (Norme matricielle). Une norme matricielle (ou
d'algèbre) est une norme sur l'espace des matrices qui vérifie une
propriété de sous-multiplicativité :
$$\|{\mathbf{A}}{\mathbf{B}}\| \leqslant \|{\mathbf{A}}\|\|{\mathbf{B}}\|$$

Les normes matricielles sont de très loin les plus intéressantes.

**Exemple 2.1** (Norme non-matricielle). $\max_{ij}a_{ij}$ est une norme
non-matricielle sur l'espace des matrices. C'est la norme $\infty$
appliquée au vecteur $m\times n$ des coefficients de
${\mathbf{A}}$ (autrement dit cette norme oublie qu'une
matrice c'est bien plus qu'un tableau de coefficients).

**Définition 2.8** (Norme subordonnée). Une norme matricielle
subordonnée est définie à partir d'une norme vectorielle :
$$\|{\mathbf{A}}\|_p=\sup_{\underset{{\mathbf{x}}\neq0}{{\mathbf{x}}\in{\mathbb{C}}^n}}\frac{\|{\mathbf{A}}{\mathbf{x}}\|_p}{\|{\mathbf{x}}\|_p}=\max_{\underset{\|{\mathbf{x}}\|_p =1}{{\mathbf{x}}\in{\mathbb{C}}^n} }\|{\mathbf{A}}{\mathbf{x}}\|_p$$
autrement dit
$\|{\mathbf{A}}{\mathbf{x}}\|_p\leqslant\|{\mathbf{A}}\|_p\|{\mathbf{x}}\|_p$,
avec au moins un vecteur où il y a égalité.





**Proposition 2.8**. *$\|{\mathbf{I}}\|_p=1$*


*Remarque 2.6*. La définition de la norme subordonnée utilise un vecteur
variant dans ${\mathbb{C}}^n$ même pour une matrice réelle.
Néanmoins, dans ce dernier cas, le vecteur qui réalise l'égalité est
réel. Bien que ce ne soit pas le cas pour les normes usuelles (1,2,
$\infty$), on peut construire une norme vectorielle et sa norme
matricielle subordonnée dont le $\max$ est différent sur
${\mathbb{C}}^n$ et ${\mathbb{R}}^n$.

Les normes $1$ et $\infty$ sont facilement calculables :
$$\|{\mathbf{A}}\|_1 = \max_{1\leqslant j\leqslant n}\sum_{i=1}^m|a_{ij}| \qquad\qquad \|{\mathbf{A}}\|_\infty = \max_{1\leqslant i\leqslant m}\sum_{j=1}^n|a_{ij}| \qquad\qquad \|{\mathbf{A}}\|_1=\|{\mathbf{A}}^T\|_\infty$$

La norme de Frobenius est matricielle, facile à calculer, mais pas
subordonnée :
$$\|{\mathbf{A}}\|_F=\sqrt{{\operatorname{tr}}({\mathbf{A}}^H{\mathbf{A}})}=\sqrt{\sum_{i=i}^m\sum_{j=1}^n|a_{ij}|^2}$$
on voit que la norme de Frobenius est en fait la norme $2$ du vecteur
composé par les colonnes de ${\mathbf{A}}$ mises bout-à-bout.
Le fait que la norme de Frobenius soit matricielle est simplement la
traduction du théorème de Cauchy-Schwarz. Cette norme (et le produit
scalaire associé) est fréquemment utilisée en calcul tensoriel
(opération de contraction).



**Proposition 2.9**. *On a les encadrements suivants :
$$\begin{aligned}
    \|{\mathbf{A}}\|_2\leqslant \|{\mathbf{A}}\|_F\leqslant\sqrt{n}\|{\mathbf{A}}\|_2\\
    \frac{1}{\sqrt{n}}\|{\mathbf{A}}\|_\infty\leqslant \|{\mathbf{A}}\|_2\leqslant\sqrt{m}\|{\mathbf{A}}\|_\infty\\
    \frac{1}{\sqrt{m}}\|{\mathbf{A}}\|_1\leqslant \|{\mathbf{A}}\|_2\leqslant\sqrt{n}\|{\mathbf{A}}\|_1
    \end{aligned}$$ et les bornes suivantes : $$\begin{aligned}
    \|{\mathbf{A}}\|_2 \leqslant \sqrt{\|{\mathbf{A}}\|_1\|{\mathbf{A}}\|_\infty}\\
    \max_{i,j}|a_{ij}|\leqslant \|{\mathbf{A}}\|_2 \leqslant \sqrt{mn}\max_{i,j}|a_{ij}|
    \end{aligned}$$*



**Proposition 2.10**. *La norme 2 est égale à la plus grande valeur
singulière de la matrice
$\|{\mathbf{A}}\|_2=\sigma_1=\sqrt{\rho({\mathbf{A}}^H{\mathbf{A}})}=\sqrt{\rho({\mathbf{A}}{\mathbf{A}}^H)}$,*

*la norme de Frobenius est égale à la racine de la somme des carrés des
valeurs singulières
$\|{\mathbf{A}}\|_F=\sqrt{\sum_i\sigma_i^2}$.*

*Si ${\mathbf{A}}$ est normale on a
$\|{\mathbf{A}}\|_2=\rho({\mathbf{A}})=\max(|lambda_i|)$,*

*en particulier si ${\mathbf{A}}$ est unitaire on a
$\|{\mathbf{A}}\|_2=1$.*

*La norme 2 et la norme de Frobenius sont inchangées par multiplication
par une matrice unitaire (les transformations unitaires sont des
isométries au sens de ces deux normes).*

**Proposition 2.11**. *Pour toute norme matricielle
$\rho(A)\leqslant \|{\mathbf{A}}\|$.*

*Etant donnée une matrice ${\mathbf{A}}$ et un nombre
$\varepsilon>0$ il existe au moins une norme subordonnée telle que
$\|{\mathbf{A}}\| \leqslant \rho(A) + \varepsilon$.*


*Preuve* Soit ${\mathbf{p}}\neq 0$ un vecteur tel que
${\mathbf{A}}{\mathbf{p}}=\lambda{\mathbf{p}}$
avec $|\lambda|=\rho({\mathbf{A}})$ et soit
${\mathbf{q}}$ un vecteur non nul alors la matrice
${\mathbf{p}}{\mathbf{q}}^H$ est non nulle et
$\rho({\mathbf{A}})\|{\mathbf{p}}{\mathbf{q}}^H\|=\|\lambda{\mathbf{p}}{\mathbf{q}}^H\|=\|{\mathbf{A}}{\mathbf{p}}{\mathbf{q}}^H\|\leqslant\|{\mathbf{A}}\|\|{\mathbf{p}}{\mathbf{q}}^H\|$.

Soit ${\mathbf{A}}$ une matrice et ${\mathbf{U}}$
une matrice inversible telle que
${\mathbf{T}}={\mathbf{U}}^{-1}{\mathbf{A}}{\mathbf{U}}$
soit triangulaire supérieure, les valeurs propres apparaissent sur la
diagonale. On définit
${\mathbf{D}}_\delta={\operatorname{diag}}{1,\delta,\delta^2,\ldots,\delta^{n-1}}$
alors
$({\mathbf{U}}{\mathbf{D}}_\delta)^{-1}{\mathbf{A}}({\mathbf{U}}{\mathbf{D}}_\delta)$
a la même diagonale que ${\mathbf{T}}$ et les coefficients
extra diagonaux multipliés par une puissance de $\delta$ ($\delta^j$ où
$j$ est le numéro de la surdiagonale), on peut choisir $\delta$ tel que
$\sum_{j=i+1}^n|\delta^{j-1}t_{ij}|\leqslant\epsilon$ pour
$1\leqslant i<n$, et alors
${\mathbf{b}}\mapsto\|{\mathbf{b}}\|=\|({\mathbf{U}}{\mathbf{D}}_\delta)^{-1}{\mathbf{B}}({\mathbf{U}}{\mathbf{D}}_\delta)\|_\infty$
répond à la question, c'est la norme subordonnée à la norme vectorielle
${\mathbf{v}}\mapsto\|({\mathbf{U}}{\mathbf{D}}_\delta)^{-1}{\mathbf{v}}\|_\infty$. ◻



**Proposition 2.12**. *Soit ${\mathbf{B}}$ une matrice
carrée. On a les équivalences suivantes*

*$\lim_{k\to \infty} {\mathbf{B}}^k =0$,*

*$\lim_{k\to \infty} {\mathbf{B}}^k {\mathbf{v}} =0$
pour tout vecteur ${\mathbf{v}}$,*

*$\rho({\mathbf{B}})<1$,*

*$\|{\mathbf{B}}\|<1$ pour au moins une norme subordonnée.*



**Proposition 2.13**. *Soit ${\mathbf{B}}$ une matrice
carrée, pour toute norme matricielle on a :
$\lim_{k\to \infty} \|{\mathbf{B}}^k\|^{1/k} =\rho({\mathbf{B}})$*


*Preuve* On a
$\rho({\mathbf{B}})\leqslant\|{\mathbf{B}}\|$ et
comme
$\rho({\mathbf{B}}^k)^{1/k}=\rho({\mathbf{B}})$,
on a
$\rho({\mathbf{B}})\leqslant\|{\mathbf{B}}^k\|^{1/k}$.
Soit
${\mathbf{B}}_\epsilon=\frac{{\mathbf{B}}}{\rho({\mathbf{B}})+\epsilon}$,
on a $\rho({\mathbf{B}}_\epsilon)<1$ donc
${\mathbf{B}}_\epsilon^k\to 0$ et donc il existe un entier
$l$ tel que pour $k\geqslant n$
$\|{\mathbf{B}}_\epsilon\|= \frac{\|{\mathbf{B}}^k\|}{(\rho({\mathbf{B}})+\epsilon)^k}\leqslant 1$
ce qui donne
$\|{\mathbf{B}}^k\|^{1/k}\leqslant \rho({\mathbf{B}})+\epsilon$
et qui permet de conclure. ◻

**Proposition 2.14**. *La norme de ${\mathbf{B}}$ donne une
information sur l'inversibilité du système perturbé :*

*soit une norme subordonnée telle que $\|{\mathbf{B}}\|<1$
alors $({\mathbf{I}}-{\mathbf{B}})$ est
inversible, on a le développement de Neumann
$({\mathbf{I}}-{\mathbf{B}})^{-1}=\sum_{k=0}^\infty {\mathbf{B}}^k$
et
$\|({\mathbf{I}}-{\mathbf{B}})^{-1}\|\leqslant \frac{1}{1-\|{\mathbf{B}}\|}$,*

*si $({\mathbf{I}}-{\mathbf{B}})$ est singulier
alors nécessairement $\|{\mathbf{B}}\|\geqslant 1$ pour toute
norme matricielle.*


*Preuve*
$({\mathbf{I}}-{\mathbf{B}}){\mathbf{u}}=0 \Rightarrow \|u\|=\|{\mathbf{B}}{\mathbf{u}}\|$
mais si $\|{\mathbf{B}}\|<1$ et
${\mathbf{u}}\neq 0$ alors
$\|{\mathbf{B}}{\mathbf{u}}\|\leqslant\|{\mathbf{B}}\|\|{\mathbf{u}}\|<\|{\mathbf{u}}\|$
on a donc nécessairement ${\mathbf{u}}=0$ et donc
$({\mathbf{I}}-{\mathbf{B}})$ est inversible. On a
alors $$\begin{aligned}
    ({\mathbf{I}}-{\mathbf{B}})^{-1}&={\mathbf{I}}+{\mathbf{B}}({\mathbf{I}}-{\mathbf{B}})^{-1}\\
    \|({\mathbf{I}}-{\mathbf{B}})^{-1}\|&\leqslant 1+\|{\mathbf{B}}\|\|({\mathbf{I}}-{\mathbf{B}})^{-1}\|\\
    \|({\mathbf{I}}-{\mathbf{B}})^{-1}\|&\leqslant \frac{1}{1-\|{\mathbf{B}}\|}
    \end{aligned}$$

$({\mathbf{I}}-{\mathbf{B}})$ singulier revient à
dire que $1$ est valeur propre de ${\mathbf{B}}$ donc
$\|{\mathbf{B}}\|\geqslant\rho({\mathbf{B}})\geqslant 1$. ◻



**Proposition 2.15** (Développement en série entière). *Soit une série
entière $\sum \alpha_k x^k$ de rayon de convergence $R$, si une matrice
carrée ${\mathbf{A}}$ a un rayon spectral
$\rho({\mathbf{A}})$ strictement inférieur à $R$ alors la
série entière de matrice $\sum \alpha_k A^k$ converge. Cette proposition
généralise le résultat sur le développement de Neumann.*
